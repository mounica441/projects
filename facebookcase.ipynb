# Data Preprocessing


```python
# label for data
from sklearn.preprocessing import LabelEncoder 
```

```python
le = LabelEncoder()
```

```python
data['Type']=le.fit_transform(data['Type'])
data['Category']=le.fit_transform(data['Category'])
data['Post Month']=le.fit_transform(data['Post Month'])
data['Post Weekday']=le.fit_transform(data['Post Weekday'])
data['Post Hour']=le.fit_transform(data['Post Hour'])
data['Lifetime Post Consumers']=le.fit_transform(data['Lifetime Post Consumers'])
data['Lifetime Post reach by people']=le.fit_transform(data['Lifetime Post reach by people'])
data['comment']=le.fit_transform(data['comment'])
data['like']=le.fit_transform(data['like'])
data['share']=le.fit_transform(data['share'])
```
